```python
# 使用日期字符串创建日期格式数据：
import numpy as np
import datetime
def strtodate(date='2020-03-03',time='00:01:02',delay_range=(10,30)):
    delay=int(np.random.randint(delay_range[0], delay_range[1], 1))
    dateTime_p = datetime.datetime.strptime(date+' '+time,'%Y-%m-%d %H:%M:%S')
    dateTime_p += datetime.timedelta(seconds=delay)
    return  dateTime_p.strftime('%Y-%m-%d'),dateTime_p.strftime('%H:%M:%S')
```
